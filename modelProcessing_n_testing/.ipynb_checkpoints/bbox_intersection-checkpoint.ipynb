```python
import numpy as np

import cv2

import matplotlib.pyplot as plt

from shapely.geometry import Polygon

import ipywidgets as wdg  # Using the ipython notebook widgets

import IPython.display as Disp
```

```python
def poly_intersection_shapely(pt1,pt2):
    """
    bool : return True or False
    """
    p1 = Polygon(pt1)
    p2 = Polygon(pt2)
    return p1.intersects(p2)
    
```

```python
def poly_intersection_mask(img,pt1,pt2):
    """
    returns area masked
    """
    img_mask = np.zeros(img.shape[:2])
    mask1 = cv2.fillConvexPoly(img_mask,np.array(pt1),1)
    mask2 = cv2.fillConvexPoly(img_mask,np.array(pt2),1)
    intersection = cv2.bitwise_and(mask1,mask2)
    return (intersection.sum()/mask1.sum(),intersection.sum()/mask2.sum())
```

```python
def poly_img(img,pts):
    pts = np.array(pts, np.int32)
    pts = pts.reshape((-1,1,2))
    cv2.polylines(img,[pts],True,(np.random.randint(0,255),np.random.randint(0,255),np.random.randint(0,255)),7)
    return img
```
