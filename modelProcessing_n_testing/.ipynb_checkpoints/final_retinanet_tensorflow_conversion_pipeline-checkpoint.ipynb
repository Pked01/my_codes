## Step 3. Checking the input/output tensor names in tensorflow model 

```python
def display_tensor_variable_names(tensorflow_model_path):
    detection_graph = tf.Graph()
    all_tensor_names = {}
    with detection_graph.as_default():
        od_graph_def = tf.GraphDef()

        with tf.gfile.GFile(tensorflow_model_path, 'rb') as fid:
            serialized_graph = fid.read()
            od_graph_def.ParseFromString(serialized_graph)
            tf.import_graph_def(od_graph_def, name='')

        ops = tf.get_default_graph().get_operations()
        all_tensor_names = {output.name for op in ops for output in op.outputs}

    for i in all_tensor_names:
        print(i)
```

## Step 4. Running inference model for prediction 

```python
config = tf.ConfigProto()
config.gpu_options.allow_growth=True
sess = tf.Session(config=config)
```

```python
def cap_list(channels_list):
    return [cv2.VideoCapture(i) for i in channels_list]
def cap_read_list(cap_list):
    res = []
    for cap in cap_list:
        ret,frame = cap.read()
        res.append((ret,frame))
    return res
        
        
```

```python
def run_inference(video_channel, class_mapping, tensorflow_model_path, write_video=False, out_video_name='inf_output.mp4'):
    
    detection_graph = tf.Graph()
    with detection_graph.as_default():
        od_graph_def = tf.GraphDef()
        with tf.gfile.GFile(tensorflow_model_path, 'rb') as fid:
            serialized_graph = fid.read()
            od_graph_def.ParseFromString(serialized_graph)
            tf.import_graph_def(od_graph_def, name='')
    
    input_tensor_name = 'input_1:0'
    output_tensor_name_list = ['filtered_detections/map/TensorArrayStack/TensorArrayGatherV3:0', 
                   'filtered_detections/map/TensorArrayStack_1/TensorArrayGatherV3:0', 
                   'filtered_detections/map/TensorArrayStack_2/TensorArrayGatherV3:0']
    
    _boxes, _scores, _labels = output_tensor_name_list
    
    labels_map= pd.read_csv(class_mapping,header=None)
    labels_to_names = dict(zip(labels_map[1],labels_map[0]))
    
    font = cv2.FONT_HERSHEY_SIMPLEX
    def nothing(x):
        pass
    bbox_colors = [(0,0,0),(0,255,0),(255,0,0),(0,255,0),(255,0,0)]
    
    cap = cv2.VideoCapture(video_channel)
    if write_video:
        out = cv2.VideoWriter(out_video_name, cv2.VideoWriter_fourcc('M','J','P','G'), 5, 
                              (int(cap.get(3)),int(cap.get(4))))
        
    cv2.namedWindow('preview', cv2.WINDOW_NORMAL)
    
    
    # create trackbars for color change
    cv2.createTrackbar('threshold','preview',0,100,nothing)
    with detection_graph.as_default():
        with tf.Session(graph=detection_graph,config=config) as sess:
            while True:
                Disp.clear_output(wait=True)
                for i in range(6):
                    ret, image = cap.read()
                
                if  not ret:
                    break
                #image = cv2.resize(image,(640,480))
                #image =  cv2.resize(image,(0,0),fx=.5,fy=.5)
                #image = cv2.resize(image,(0,0),fx=.5,f)
                # the array based representation of the image will be used later in order to prepare the
                # result image with boxes and labels on it.
                #image_np = load_image_into_numpy_array(image)
                # Expand dimensions since the model expects images to have shape: [1, None, None, 3]
                image_np_expanded = np.expand_dims(image, axis=0)

                # image_tensor = detection_graph.get_tensor_by_name('input_1:0')
                image_tensor = detection_graph.get_tensor_by_name(input_tensor_name)
                # Each box represents a part of the image where a particular object was detected.

                # boxes = detection_graph.get_tensor_by_name('filtered_detections/map/TensorArrayStack/TensorArrayGatherV3:0')
                boxes = detection_graph.get_tensor_by_name(_boxes)
                # Each score represent how level of confidence for each of the objects.
                
                # Score is shown on the result image, together with the class label.
                # scores = detection_graph.get_tensor_by_name('filtered_detections/map/TensorArrayStack_1/TensorArrayGatherV3:0')
                scores = detection_graph.get_tensor_by_name(_scores)
                
                # labels = detection_graph.get_tensor_by_name('filtered_detections/map/TensorArrayStack_2/TensorArrayGatherV3:0')
                labels = detection_graph.get_tensor_by_name(_labels)
                # num_detections = detection_graph.get_tensor_by_name('num_detections:0')
                # Actual detection.
                # process image
                start = time.time()
                (boxes, scores, labels) = sess.run(
                  [boxes, scores,labels],
                  feed_dict={image_tensor: image_np_expanded})
                
                # copy to draw on
                draw = image.copy()
                draw = cv2.cvtColor(draw, cv2.COLOR_BGR2RGB)
                
                # NOT USING BATCH PROCESSING 
                #boxes, scores, labels = model.predict_on_batch(np.array([image,image]))
                #boxes,scores,labels = boxes[0],scores[0],labels[0]
                print("FPS : ", 1/(time.time() - start))

                # correct for image scale
                #boxes /= scale

                # visualize detections
                for box, score, label in zip(boxes[0], scores[0], labels[0]):
                    # scores are sorted so we can break
                    thresh_score = cv2.getTrackbarPos('threshold','preview')
                    if score < .01*thresh_score:
                        break
                    
                    color = bbox_colors[(label)]
                    b = box.astype(int)
                    draw_box(draw, b, color=color)
                    caption = "{} {:.3f}".format(labels_to_names[label], score)
                    draw_caption(draw, b, caption)
                
                cv2.putText(draw,"threshold = "+str(.01*thresh_score),(draw.shape[1]-300,40), font, 1,(0,0,255),2,cv2.LINE_AA)
                draw_bgr = cv2.cvtColor(draw,cv2.COLOR_RGB2BGR)
                
                if write_video:
                    out.write(draw_bgr)

                cv2.imshow("preview", draw_bgr)
                k = cv2.waitKey(1)
                if k==27:
                    break
            if write_video:
                out.release()
            #tf.reset_default_graph()
            cv2.destroyAllWindows()
            cap.release()
            
        
        # Visualization of the results of a detection. - USING VISUALIZATION UTIL
        #         vis_util.visualize_boxes_and_labels_on_image_array(
        #           image_np,
        #           np.squeeze(boxes),
        #           np.squeeze(classes).astype(np.int32),
        #           np.squeeze(scores),
        #           category_index,
        #           use_normalized_coordinates=True,
        #           line_thickness=8)
```
